```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("fivethirtyeight")
import warnings # 경고 메시지 무시
warnings.filterwarnings(action='ignore')
from tqdm import tqdm # 작업 프로세스 시각화
from nltk import word_tokenize, bigrams #토큰화, 바이그램
from nltk.util import ngrams #트라이그램
from nltk import ConditionalFreqDist #동시출현빈도 계수화
from konlpy.tag import Mecab
import re # 문자열 처리를 위한 정규표현식 패키지
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
df = pd.read_csv('/content/drive/MyDrive/suncare_rev_p2.csv', index_col=0)
```

```python
df.head()
```

```python
text = df[['review']]
text.head()
```

```python
text.shape
```

```python
# 텍스트화된 리뷰 전처리
def RevProcess(rev):
    rev = rev.replace(".", "").strip()
    rev = rev.replace(":", "").strip()
    rev = re.sub('[^가-힣]', ' ', rev)
    rev = rev.replace("  "," ").replace("   ", " ").replace("    "," ").replace("     ", " ").replace("       ", " ")
    return rev
```

```python
text['review'] = text.apply(lambda x: RevProcess(x['review']), axis=1)
```

```python
text['review']
```

```python
tagger = Mecab()
results = []
for post in text["review"]:
  word_tokens = tagger.pos(post)
  result = [s for s, t in word_tokens if t in ['NNG','XR', 'VA'] and len(s)>1]
  results.append(result)
```

```python
len(results)
```

```python
text['keyword'] = results
```

```python
word_df = pd.read_csv('/content/drive/MyDrive/suncare_total_2 (1).xlsx - Sheet1.csv')
```

```python
stopwords = word_df[word_df['태그1'] == 's']['형태소/개체명'].values.tolist()
```

```python
for i in stopwords:
  for word in text['keyword']:
    if word == i:
      text['keyword'] = text['keyword'].apply(lambda x: x.replace(i,''))
```

```python
for i,j in zip(word_df['형태소/개체명'], word_df['태그1']):
  for word in text['keyword']:
    if i in word:
      text['keyword'] = text['keyword'].apply(lambda x: x.replace(i,j))
    else:
      pass
```

```python
text['keyword'][:5]
```

```python
vectorizer = TfidfVectorizer(stop_words=stopwords)
X = vectorizer.fit_transform(text['keyword'])

print('TF-IDF 행렬의 크기 :', X.shape)
```

```python
lda_model = LatentDirichletAllocation(n_components=10,learning_method='online',random_state=1,max_iter=1)
```

```python
lda_top = lda_model.fit_transform(X)
```

```python
print(lda_model.components_)
print(lda_model.components_.shape)
```

```python
# 단어 집합. 1,000개의 단어가 저장됨.
terms = vectorizer.get_feature_names()

def get_topics(components, feature_names, n=5):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1), [(feature_names[i], topic[i].round(2)) for i in topic.argsort()[:-n - 1:-1]])

get_topics(lda_model.components_,terms)
```
