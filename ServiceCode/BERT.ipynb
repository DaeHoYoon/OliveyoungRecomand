```python
# 텍스트화된 리뷰 전처리
def RevProcess(rev):
    rev = rev.replace(".", "").strip()
    rev = rev.replace(":", "").strip()
    rev = re.sub('[^가-힣]', ' ', rev)
    rev = rev.replace("  "," ").replace("   ", " ").replace("    "," ").replace("     ", " ").replace("       ", " ")
    return rev
```

```python
class CustomTokenizer:
    def __init__(self, tagger):
        self.tagger = tagger
    def __call__(self, sent):
        word_tokens = self.tagger.pos(sent)
        result = [s for s, t in word_tokens if t in ['NNG','XR'] and len(s)>1 and s not in stopwords]
        return result
```

```python
sentence = []
for rev in tqdm(df['review']):
    sent = RevProcess(str(rev))
    # sent = kss.split_sentences(sent)
    sentence.append(sent)
```

```python
sentence
```

```python
# 문장에서 단어 사이의 관계 확인
bigram = gensim.models.Phrases(sentence) # 두 글자 사이의 관계 확인
# trigram = gensim.models.Phrases(bigram[sentence]) # 세 글자 사이의 관계 확인
bigram_mod = gensim.models.phrases.Phraser(bigram)
# trigram_mod = gensim.models.phrases.Phraser(trigram)

bigram_document = [bigram_mod[doc] for doc in sentence]
id2word = corpora.Dictionary(bigram_document) # 단어에 레이블링 및 빈도수 기록
corpus = [id2word.doc2bow(doc) for doc in bigram_document] # 문서데이터를 수치화한 값을 리스트로 담는다

coherence_values = []
for i in tqdm(range(2,11)):
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, id2word=id2word, num_topics=i, random_state=1)
    coherence_model_lda = CoherenceModel(model = ldamodel, texts=bigram_document, dictionary=id2word, topn=10)
    coherence_lda = coherence_model_lda.get_coherence()
    coherence_values.append(coherence_lda)
    print('n:', i, 'Coherence Score:', coherence_lda)
```

```python
x = range(2, 11)

plt.figure(figsize=(15, 7))
plt.plot(x, coherence_values)
plt.title('Coherence Values', size=25)
plt.xlabel('Number of Topics')
plt.xlim(1, 10)
plt.xticks(x, range(2, 10), fontsize=15)

plt.ylabel('Coherence Score')
plt.yticks(fontsize=15)
# plt.grid()
plt.show()
```

```python
customTokenizer = CustomTokenizer(Mecab())
vectorizer = CountVectorizer(tokenizer=customTokenizer)
train_bow_embeddings = vectorizer.fit_transform(sentence)
print(train_bow_embeddings.shape)
```

```python
vocab = vectorizer.get_feature_names()
id2token = {k: v for k, v in zip(range(0, len(vocab)), vocab)}
len(vocab)
```

```python
train_contextualized_embeddings = bert_embeddings_from_list(sentence,"sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens",  max_seq_length=128)
```

```python
qt = TopicModelDataPreparation()
training_dataset = qt.load(train_contextualized_embeddings, train_bow_embeddings, id2token)
```

```python
ctm = CombinedTM(bow_size=len(vocab), contextual_size=768, n_components=5, num_epochs=3)
ctm.fit(training_dataset)
```

```python
ctm.get_topics(10)
```

```python
lda_vis_data = ctm.get_ldavis_data_format(vocab, training_dataset, n_samples=10)
ctm_pd = vis.prepare(**lda_vis_data)
vis.display(ctm_pd)
```
